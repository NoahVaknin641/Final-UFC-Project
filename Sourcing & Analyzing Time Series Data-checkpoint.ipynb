```python
# Importing packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import statsmodels.api as sm
import warnings

warnings.filterwarnings("ignore")

plt.style.use('fivethirtyeight')
```

```python
path = r'C:\Users\nvak6\CareerFoundry UFC Project'
UFC_data = pd.read_csv(os.path.join(path, 'UFC_Cleaned.csv'))
```

```python
UFC_data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
UFC_data.head(1)
```

```python
UFC_data_2 = UFC_data.reset_index()
```

```python
UFC_data_2['Fight_Date'] = pd.to_datetime(UFC_data_2['Fight_Date'], format='%m/%d/%Y')
UFC_data_sub = UFC_data_2.loc[(UFC_data_2['Fight_Date'] >= '2011-01-01') & (UFC_data_2['Fight_Date'] < '2021-03-21')]
```

Limiting the analysis to 2011–2021 allows you to focus on a period with better data consistency, relevant trends, and a more comparable group of fighters. This makes your analysis more focused and likely to yield more meaningful insights, especially if you're interested in understanding current dynamics within the UFC.

```python
UFC_data_sub = UFC_data_sub.dropna(subset=['Red_Avg_Sig_Strike_Percent', 'Blue_Avg_Sig_Strike_Percent'])
```

```python
# Setting the date as the index

from datetime import datetime

UFC_data_sub['datetime'] = pd.to_datetime(UFC_data_sub['Fight_Date']) # create datetime index from Fight_Date column
UFC_data_sub = UFC_data_sub.set_index('datetime') # set the datetime as the index of the df
UFC_data_sub.drop(['Fight_Date'], axis = 1, inplace = True) # drop the Fight_Date column
UFC_data_sub.tail()
                                          
```

```python
UFC_red = UFC_data_sub['Red_Avg_Sig_Strike_Percent']
```

```python
UFC_red_avg = UFC_red.groupby(UFC_data_sub.index).mean()
```

```python
UFC_red_avg.head()
```

```python
plt.figure(figsize=(15,5), dpi=100)
plt.plot(UFC_red_avg)
```

```python
decomposition_red = sm.tsa.seasonal_decompose(UFC_red_avg, model='additive', period=7)
```

```python
# Set the plot size to be as large as possible
plt.figure(figsize=(24, 14))  # Large frame (24 inches wide by 14 inches tall)

# Plot the decomposition
decomposition_red.plot()

# Adjust the layout to ensure proper spacing and prevent overlap
plt.tight_layout()

# Add a title for clarity
plt.suptitle('Seasonal Decomposition of Red Average Significant Strike Percent', fontsize=22, y=1.05)

# Display the plot
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller # Import the adfuller() function

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)
```

```python
dickey_fuller(UFC_red_avg)
```

The time series is definitively stationary after accounting for trend and seasonal components. The very large negative test statistic and zero p-value provide strong statistical evidence that the underlying process is stable and mean-reverting, despite any apparent trends we might see in the visual plot. This suggests that changes in the series tend to revert to a baseline level over time.

```python
dickey_fuller(UFC_blue_avg)
```

This shows strong evidence of stationarity in the time series after accounting for trend and seasonal components. The test statistic is decisively more negative than the critical values, and the extremely small p-value provides very strong evidence against non-stationarity. This indicates that the series has stable statistical properties and tends to revert to its mean level despite any apparent trends.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(UFC_red_avg)
plt.show()
```

This autocorrelation plot suggests that the time series is well-behaved and shows little to no serial correlation. The lack of significant correlations at any lag indicates that past values are not strong predictors of future values, which is consistent with an efficient or random process. This supports the findings from the earlier Dickey-Fuller test about the series being stationary.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(UFC_blue_avg)
plt.show()
```

The plot indicates that observations in this time series are largely independent of each other at different time lags, suggesting a well-behaved, stationary process with little to no serial correlation. This means past values don't have significant predictive power for future values.
