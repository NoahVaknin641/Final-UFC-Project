# 6.5: Unsupervised Machine Learning: Clustering

```python
# Importing Packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.cluster import KMeans
import pylab as pl
```

```python
path = r'C:\Users\nvak6\CareerFoundry UFC Project'
UFC_data = pd.read_csv(os.path.join(path, 'UFC_Cleaned.csv'))
```

```python
# Check the data types and missing values
UFC_data.info()
UFC_data.isnull().sum()
```

```python
# Remove categorical columns (keep only numerical ones)
UFC_data_cleaned = UFC_data.select_dtypes(include=[np.number])
UFC_data_cleaned = UFC_data_cleaned.dropna()
```

```python
num_cl = range(1,10)
kmeans = [KMeans(n_clusters = i) for i in num_cl]
```

```python
score = [kmeans[i].fit(UFC_data_cleaned ).score(UFC_data_cleaned ) for i in range(len(kmeans))]
```

```python
score
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(UFC_data_cleaned)
```

```python
UFC_data_cleaned['clusters'] = kmeans.fit_predict(UFC_data_cleaned)
```

```python
UFC_data_cleaned.head()
```

```python
UFC_data_cleaned['clusters'].value_counts()
```

```python
# Choose variables you want to plot
var1 = 'Red_Avg_Sig_Strikes_Landed'
var2 = 'Blue_Avg_Sig_Strikes_Landed'

# Create a scatter plot for the two variables, coloring by the clusters
plt.figure(figsize=(10, 6))
plt.scatter(UFC_data_cleaned[var1], UFC_data_cleaned[var2], c=UFC_data_cleaned['clusters'], cmap='viridis')
plt.title(f'Clusters based on {var1} and {var2}')
plt.xlabel(var1)
plt.ylabel(var2)
plt.colorbar(label='Cluster')
plt.show()

# Display the first few rows of the data with the selected variables and cluster assignments
print(UFC_data_cleaned[[var1, var2, 'clusters']].head())
```

The scatter plot displays the relationship between "Red_Avg_Sig_Strikes_Landed" and "Blue_Avg_Sig_Strikes_Landed" for a dataset, with the data points colored by cluster assignments.
Key observations:

The data points exhibit a positive correlation, with higher values on the x-axis (Red_Avg_Sig_Strikes_Landed) corresponding to higher values on the y-axis (Blue_Avg_Sig_Strikes_Landed).
The data is grouped into distinct clusters, indicated by the color gradient from blue to yellow to red.
The rightmost cluster contains the data points with the highest combination of "Red_Avg_Sig_Strikes_Landed" and "Blue_Avg_Sig_Strikes_Landed" values.

```python
# Group by the clusters and calculate descriptive statistics for each group
cluster_stats = UFC_data_cleaned.groupby('clusters').describe()
```
