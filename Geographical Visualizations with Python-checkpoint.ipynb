```python
#Importing Cleaned UFC Data
path = r'C:\Users\nvak6\CareerFoundry UFC Project'
UFC_data = pd.read_csv(os.path.join(path, 'UFC_Cleaned.csv'))
```

```python
# Aggregate the number of events by Event_Location
event_counts = UFC_data.groupby('Event_Location').size().reset_index(name='Event_Count')
```

```python
# Geolocator for getting latitude and longitude from location names
geolocator = Nominatim(user_agent="event_locator")

# Function to get latitude and longitude from location
def get_location_coordinates(location):
    try:
        geoloc = geolocator.geocode(location)
        if geoloc:
            return geoloc.latitude, geoloc.longitude
        else:
            return None, None
    except Exception as e:
        return None, None

# Add latitude and longitude columns to the event_counts DataFrame
event_counts['Latitude'], event_counts['Longitude'] = zip(*event_counts['Event_Location'].map(get_location_coordinates))

# Drop rows where coordinates are not available (to avoid errors in map plotting)
event_counts = event_counts.dropna(subset=['Latitude', 'Longitude'])

# Create a base map
m = folium.Map(location=[20, 0], zoom_start=2)

# Add markers for each event location based on count
for _, row in event_counts.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5 + row['Event_Count'] // 10,  # Make marker size proportional to the event count
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"{row['Event_Location']}: {row['Event_Count']} events"
    ).add_to(m)
```
