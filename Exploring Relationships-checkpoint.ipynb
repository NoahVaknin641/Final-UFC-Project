# 6.2: Exploring Relationships

```python
#Importing Packages
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#Importing Cleaned UFC Data
path = r'C:\Users\nvak6\CareerFoundry UFC Project'
UFC_data = pd.read_csv(os.path.join(path, 'UFC_Cleaned.csv'))
```

## How do fighter statistics (damage dealt, damage received, etc.) correlate with the likelihood of winning a fight?

```python
# Assign 1 if the red fighter won, 0 if the blue fighter won
UFC_data['Winner_Encoded'] = UFC_data['Winner'].apply(lambda x: 1 if x == 'Red' else 0)
```

```python
# Define metric groups
metric_groups = {
    "Striking Efficiency and Output": [
        'Red_Avg_Sig_Strike_Percent', 'Blue_Avg_Sig_Strike_Percent',
        'Red_Avg_Sig_Strikes_Landed', 'Blue_Avg_Sig_Strikes_Landed',
        'Winner_Encoded'
    ],
    "Control Metrics": [
        'Red_Avg_Control_Time_Seconds', 'Blue_Avg_Control_Time_Seconds',
        'Red_Avg_Takedowns_Landed', 'Blue_Avg_Takedowns_Landed',
        'Winner_Encoded'
    ],
    "Finishing Ability": [
        'Red_Avg_Knockdowns', 'Blue_Avg_Knockdowns',
        'Red_Avg_Submission_Attempts', 'Blue_Avg_Submission_Attempts',
        'Winner_Encoded'
    ],
    "Momentum": [
        'Red_Current_Win_Streak', 'Blue_Current_Win_Streak',
        'Red_Longest_Win_Streak', 'Blue_Longest_Win_Streak',
        'Winner_Encoded'
    ]
}

# Generate heatmaps for each group
for group_name, metrics in metric_groups.items():
    # Filter DataFrame to include only relevant columns
    group_data = UFC_data[metrics]
    
    # Calculate correlation matrix
    correlation_matrix = group_data.corr()
    
    # Plot heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'{group_name} Correlation Heatmap')
    plt.show()
```

From these heat maps it seems that no one stat that we have looked at is soley responsible for the outcome of the fight

## What are the most significant performance metrics that influence a fighter’s chances of winning (e.g., strikes landed, significant strikes, takedowns)

```python
# Convert the 'Winner' column into binary (1 for Blue wins, 0 for Red wins)
UFC_data['Winner'] = UFC_data['Winner'].apply(lambda x: 1 if x == 'Blue' else 0)

# Select performance metrics for scatter plot (e.g., significant strikes, takedowns)
metrics = ['Blue_Avg_Sig_Strikes_Landed', 'Blue_Avg_Takedowns_Landed']

# Create scatter plots for these metrics
for metric in metrics:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=UFC_data, x=metric, y='Winner', alpha=0.6, palette='deep')
    plt.title(f'Scatter Plot of {metric} vs. Fight Outcome (Winner)')
    plt.xlabel(f'{metric}')
    plt.ylabel('Winner (0 = Red, 1 = Blue)')
    plt.show()
```

From these two scatterplotts, it seems that takedowns have a bigger outcome on winning a fight than significant strikes.

## How do fighter demographics (age, weight class, reach, etc.) correlate with the fight outcome?

```python
# Select demographic columns (including age, height, weight, reach for both fighters)
demographics = ['Blue_Age', 'Blue_Weight_Lbs', 'Blue_Height_Cms', 'Blue_Reach_Cms', 'Red_Age', 'Red_Weight_Lbs', 'Red_Height_Cms', 'Red_Reach_Cms']

# Create a new DataFrame with the selected demographics and the winner column
df_demographics = UFC_data[demographics + ['Winner']]

# Create a pair plot to visualize correlations between fighter demographics and the fight outcome
sns.pairplot(df_demographics, hue='Winner', palette={0: 'red', 1: 'blue'}, plot_kws={'alpha': 0.6})

# Display the plot
plt.suptitle('Fighter Demographics vs Fight Outcome', y=1.02)
plt.show()
```

From these plots it seems height and reach are the most impactful

## What are the most common outcomes in UFC fights (e.g., KO, submission, decision)?

```python
# Extract relevant columns from the UFC_data DataFrame
win_metrics_columns = [
    'Blue_Wins_by_KO_TKO', 'Blue_Wins_by_Submission', 'Blue_Wins_by_Decision_Unanimous',
    'Blue_Wins_by_Decision_Split', 'Blue_Wins_by_Decision_Majority', 'Red_Wins_by_KO_TKO',
    'Red_Wins_by_Submission', 'Red_Wins_by_Decision_Unanimous', 'Red_Wins_by_Decision_Split',
    'Red_Wins_by_Decision_Majority'
]

# Create a DataFrame focusing on the relevant columns
win_metrics_df = UFC_data[win_metrics_columns]

# Calculate the total wins for blue and red fighters
win_metrics_df['Blue_Total_Wins'] = win_metrics_df[['Blue_Wins_by_KO_TKO', 'Blue_Wins_by_Submission', 'Blue_Wins_by_Decision_Unanimous', 
                                                     'Blue_Wins_by_Decision_Split', 'Blue_Wins_by_Decision_Majority']].sum(axis=1)

win_metrics_df['Red_Total_Wins'] = win_metrics_df[['Red_Wins_by_KO_TKO', 'Red_Wins_by_Submission', 'Red_Wins_by_Decision_Unanimous', 
                                                    'Red_Wins_by_Decision_Split', 'Red_Wins_by_Decision_Majority']].sum(axis=1)

# Calculate the percentage of each win type relative to total wins for each fighter
win_metrics_df['Blue_KO_TKO_Percentage'] = win_metrics_df['Blue_Wins_by_KO_TKO'] / win_metrics_df['Blue_Total_Wins'] * 100
win_metrics_df['Blue_Submission_Percentage'] = win_metrics_df['Blue_Wins_by_Submission'] / win_metrics_df['Blue_Total_Wins'] * 100
win_metrics_df['Blue_Decision_Unanimous_Percentage'] = win_metrics_df['Blue_Wins_by_Decision_Unanimous'] / win_metrics_df['Blue_Total_Wins'] * 100
win_metrics_df['Blue_Decision_Split_Percentage'] = win_metrics_df['Blue_Wins_by_Decision_Split'] / win_metrics_df['Blue_Total_Wins'] * 100
win_metrics_df['Blue_Decision_Majority_Percentage'] = win_metrics_df['Blue_Wins_by_Decision_Majority'] / win_metrics_df['Blue_Total_Wins'] * 100

win_metrics_df['Red_KO_TKO_Percentage'] = win_metrics_df['Red_Wins_by_KO_TKO'] / win_metrics_df['Red_Total_Wins'] * 100
win_metrics_df['Red_Submission_Percentage'] = win_metrics_df['Red_Wins_by_Submission'] / win_metrics_df['Red_Total_Wins'] * 100
win_metrics_df['Red_Decision_Unanimous_Percentage'] = win_metrics_df['Red_Wins_by_Decision_Unanimous'] / win_metrics_df['Red_Total_Wins'] * 100
win_metrics_df['Red_Decision_Split_Percentage'] = win_metrics_df['Red_Wins_by_Decision_Split'] / win_metrics_df['Red_Total_Wins'] * 100
win_metrics_df['Red_Decision_Majority_Percentage'] = win_metrics_df['Red_Wins_by_Decision_Majority'] / win_metrics_df['Red_Total_Wins'] * 100

# Calculate the average percentages for each win type across all fights
avg_win_metrics = win_metrics_df[['Blue_KO_TKO_Percentage', 'Blue_Submission_Percentage', 'Blue_Decision_Unanimous_Percentage', 
                                  'Blue_Decision_Split_Percentage', 'Blue_Decision_Majority_Percentage', 
                                  'Red_KO_TKO_Percentage', 'Red_Submission_Percentage', 'Red_Decision_Unanimous_Percentage', 
                                  'Red_Decision_Split_Percentage', 'Red_Decision_Majority_Percentage']].mean()

# Reshape the data for easier plotting
avg_win_metrics_df = avg_win_metrics.reset_index().rename(columns={'index': 'Win Metric', 0: 'Average Percentage'})

# Create a barplot to visualize the average win metrics for Red and Blue fighters
plt.figure(figsize=(10, 6))
sns.barplot(x='Win Metric', y='Average Percentage', data=avg_win_metrics_df, palette='Set2')

# Add title and labels
plt.title('Average Win Percentage by Win Metric (Red vs Blue Fighters)', fontsize=16)
plt.xlabel('Win Metric', fontsize=12)
plt.ylabel('Average Percentage (%)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

From this plot it seems that most wins happen by tko
